```python
allData = pd.read_csv('PDI_Police_Calls_for_Service__30MAR2019.csv', low_memory = False)
allData.describe(include = 'all')
```

```python
allData.dropna(subset = ['LATITUDE_X','LONGITUDE_X','ARRIVAL_TIME_PRIMARY_UNIT','DISPATCH_TIME_PRIMARY_UNIT'], inplace = True)
allData.describe(include = 'all')
```

```python
allData['ARRIVAL_TIME_PRIMARY_UNIT'] = pd.to_datetime(allData['ARRIVAL_TIME_PRIMARY_UNIT'])
allData['DISPATCH_TIME_PRIMARY_UNIT'] = pd.to_datetime(allData['DISPATCH_TIME_PRIMARY_UNIT'])
allData['CREATE_TIME_INCIDENT'] = pd.to_datetime(allData['CREATE_TIME_INCIDENT'])
allData['CLOSED_TIME_INCIDENT'] = pd.to_datetime(allData['CLOSED_TIME_INCIDENT'])
allData.dtypes
```

```python
allData['ResponseTime'] = (allData['ARRIVAL_TIME_PRIMARY_UNIT'] - \
                           allData['DISPATCH_TIME_PRIMARY_UNIT']).astype('timedelta64[m]')
allData.head(100)
```

## Data Filtering

Filtering all dispatch-arrival deltas (Response Times) to those greater than 0.  The values with less than 0 are ones where there was no dispatch time but an arrival time (I suppose I could use creation time for dispatch time instead?
Will examine in future iterations)

Also dropping a few locations where I heavily suspect there is erroneous data (this was a lot of google mapping.  I am certain there are more, and will need to continue looking for lat/longs):
- All 23XX FERGUSON RD dispatches, since that is where Police District 3 headquarters is (2300 Ferguson Rd)
- All 3XX EZZARD CHARLES dispatches, since that is where another Police District headquarters is (310 Ezzard Charles)
- All 41XX READING RD dispatches, since that is where Police District 4 headquarters is (4150 Reading Rd)
- All 32XX ERIE AV dispatches, since that is where Police District 2 headquarters is(3295 Erie Av)
- All 10XX LUDLOW AV dispatches, since that is where the Police District 5 headquarters used to be until April 2018 (1012 LUDLOW AV)
- All 58XX HAMILTON AV dispatches, since that is the new Police District 5 headquarters
- All 8XX EVANS ST dispatches, since that is the location of the police academy
- All 10XX MAIN ST dispatches, since that is the courthouse
- All 20XX RADCLIFF DR dispatches, since that is where the Hamilton County Emergency Management Center is, and it looks like they use it for location testing


ALSO getting rid of all dispatch items that were 'CAN:CANCEL' and where incident_type is SDET (based on the comments that appears to be off duty police logging something that already happened.  Their times are ridiculous.

While I am not doing this now, we should debate getting rid of any DIRPAT incident types as well.  These I believe are incidents that police call in while on a 'Directed Patrol' and may not reflect an actuall 911 call
    


```python
allData = allData.drop(allData[allData.ResponseTime<= 0].index)
allData = allData.drop(allData[allData.ADDRESS_X == '23XX FERGUSON RD'].index)
allData = allData.drop(allData[allData.ADDRESS_X == '41XX READING RD'].index)
allData = allData.drop(allData[allData.ADDRESS_X == '32XX ERIE AV'].index)
allData = allData.drop(allData[allData.ADDRESS_X == '10XX LUDLOW AV'].index)
allData = allData.drop(allData[allData.ADDRESS_X == '58XX HAMILTON AV'].index)
allData = allData.drop(allData[allData.ADDRESS_X == '8XX EVANS ST'].index)
allData = allData.drop(allData[allData.ADDRESS_X == '10XX MAIN ST'].index)
allData = allData.drop(allData[allData.ADDRESS_X == '20XX RADCLIFF DR'].index)
allData = allData.drop(allData[allData.ADDRESS_X == '3XX EZZARD CHARLES DR'].index)
allData = allData.drop(allData[allData.DISPOSITION_TEXT == 'CAN:CANCEL'].index)
allData = allData.drop(allData[allData.INCIDENT_TYPE_ID == 'SDET'].index)
```

```python
allData.describe(include = 'all')
```

Let's create a few new columns that capture pertinent temporal information.  Break the dispatch time up into year, month, day, hour, minute, and the day of the week.
    
We probably need to figure out why there is a difference between the creation time and the dispatch time, and why not every event has a dispatch time (where all NaN's have already been filtered out). 


```python

allData['year'] = pd.DatetimeIndex(allData['DISPATCH_TIME_PRIMARY_UNIT']).year.astype(str)
allData['month'] = pd.DatetimeIndex(allData['DISPATCH_TIME_PRIMARY_UNIT']).month.astype(str)
allData['day'] = pd.DatetimeIndex(allData['DISPATCH_TIME_PRIMARY_UNIT']).day
allData['hour'] = pd.DatetimeIndex(allData['DISPATCH_TIME_PRIMARY_UNIT']).hour
allData['minute'] = pd.DatetimeIndex(allData['DISPATCH_TIME_PRIMARY_UNIT']).minute
allData['dayofweek'] = pd.DatetimeIndex(allData['DISPATCH_TIME_PRIMARY_UNIT']).dayofweek.astype(str)

allData.replace({'month':{
                         '0' : 'Jan',
                         '1' : 'Feb',
                         '2' : 'Mar',
                         '3' : 'Apr',
                         '4' : 'May',
                         '5' : 'Jun',
                         '6' : 'Jul',
                         '7' : 'Aug',
                         '8' : 'Sep',
                         '9' : 'Oct',
                         '10': 'Nov',
                         '11': 'Dec'
                        },
               'dayofweek':{
                           '0' : 'Mon',
                           '1' : 'Tue',
                           '2' : 'Wed',
                           '3' : 'Thu',
                           '4' : 'Fri',
                           '5' : 'Sat',
                           '6' : 'Sun'
                           }
                       }, inplace = True)


allData.describe(include='all')
```

Because of the skewed nature of the data, for graphing purposes in Dash I am going to make a 'color' column that will reflect the response time from 25th percentile to 75th percentile.  Any values that are below the 25th percentile will be replaced with the 25th percentile value, and likewise with the 75th percentile.  This column should only be used for visualizations and no predictions

```python
allData['colorScale'] = allData['ResponseTime']
upper = allData['ResponseTime'].quantile(0.75)
lower = allData['ResponseTime'].quantile(0.25)
allData['colorScale'] = np.where(allData['colorScale']>upper,upper,allData['colorScale'])
allData['colorScale'] = np.where(allData['colorScale']<lower,lower,allData['colorScale'])

allData.describe(include = 'all')
```
